```python
from paddleocr import PaddleOCR, draw_ocr
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

```python
!sudo apt-get install fonts-dejavu-core
```

```python
# Initialize PaddleOCR with English model
ocr = PaddleOCR(use_angle_cls=True, lang='en')

# Load the image
img_path = '/content/bourbon.jpg'
img = cv2.imread(img_path)

# Convert BGR to RGB for PaddleOCR
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Perform OCR
result = ocr.ocr(img)

# Print results (extracted text and confidence scores)
for res in result:
    for line in res:
        print(f"Text: {line[1][0]}, Confidence: {line[1][1]}")

# Install and set font for Colab environment
!apt-get install -y fonts-dejavu-core
font_path = '/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf'

# If you want to visualize the result
image = Image.open(img_path)
boxes = [elements[0] for elements in result[0]]
txts = [elements[1][0] for elements in result[0]]
scores = [elements[1][1] for elements in result[0]]

# Draw the results
im_show = draw_ocr(image, boxes, txts, scores, font_path=font_path)
im_show = Image.fromarray(im_show)

# Show the image with drawn OCR results
plt.imshow(im_show)
plt.show()
```
